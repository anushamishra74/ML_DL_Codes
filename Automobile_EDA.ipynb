# Exploratory Data Analysis

```python
#importing libraries
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#reading dataset
data = pd.read_csv('automobileEDA.csv')
data.head() #to get the first 5 rows
```

```python
#to get the last 5 rows
data.tail()
```

```python
#to get the dimensions of the dataset
data.shape
```

```python
#to get some additional information of the dataset
data.info()
```

```python
#to get column names
data.columns
```

```python
#to get statistical information of data
data.describe()
```

```python
#to check for NaN values
data.isna().any()
```

```python
#to check for null values
data.isnull().any()
```

```python
#to get the sum of null values
data.isnull().sum()
```

```python
#to remove null values
data.dropna(subset=['stroke','horsepower-binned'],inplace = True)
```

```python
data.isnull().sum()
```

```python
#to build correlation matrix for dataset
corr = data.corr() 
fig = plt.figure(figsize = (14, 14)) 
sns.heatmap(corr, annot= True) #cmap=plt.cm.afmhot) 
plt.title("Correlation matrix")
plt.show()
```

```python
#How many cars are there in the dataset?
data['make'].count
```

```python
#What model of cars are present in the dataset?
data['make'].unique()
```

```python
#What is highest price in the datset?
data['price'].max()
```

```python
#What is lowest price in the datset?
data['price'].min()
```

```python
#Which model of car has the highest price?
data[data['price'] == data['price'].max()]['make']
```

```python
#Which model of car has the lowest price?
data[data['price'] == data['price'].min()]['make']
```

# Data Visualization

```python
# Univariate Distribution
sns.histplot(data['engine-location']) #univariate analysis by creating histogram for string data
plt.show()
```

```python
sns.histplot(data['normalized-losses']) #univariate analysis by creating histogram
plt.show()
```

```python
sns.displot(data['normalized-losses'],kde = False,rug = True) # univariate analysis by using distplot
plt.show()
```

```python
sns.displot(data['normalized-losses']) # univariate analysis by using distplot without 'kde' and 'rug' parameters
plt.show()
```

```python
# Bivariate Analysis
#kind : {scatter, hex, reg, kde, resid}
sns.jointplot(data['engine-size'], data['horsepower'],kind='hex')
plt.show()
```

```python
# Multivariate Analysis
sns.pairplot(data[['normalized-losses','engine-size','horsepower']])
```

```python
#to plot continuous and discrete variables together
sns.stripplot(data['drive-wheels'],data['length'],jitter=True)
```

```python
sns.swarmplot(data['drive-wheels'],data['length'])
```

```python
#to plot continuous and discrete variables together
sns.boxplot(data['num-of-doors'], data['horsepower'],hue=data['drive-wheels']) #hue is applied on categorical data
```

```python
# barplot is used when there is meaningful value like 0 in temperature.
# It shows the smallest and the highest value
sns.barplot(data['body-style'],data['horsepower'])
```

```python
#special case of barplot is counplot which is used to plot the frequency of any categorical variable.
sns.countplot(data['body-style'],hue=data['engine-location'])
```

```python
#pointplot shows the highest point and also the line trend
sns.pointplot(data['fuel-system'], data['horsepower'])
```

```python
sns.pointplot(data['fuel-system'], data['horsepower'],hue=data['num-of-doors'])
```

```python
#it is used to plot multiple categorical variables. It is called multi-panel categorical plot
#The `factorplot` function has been renamed to `catplot`
#the default `kind` in `factorplot` (`'point'`) has changed `'strip'` in `catplot`.
sns.factorplot(x='drive-wheels',
               y='price',
               hue='num-of-doors',
               col='engine-location',
               data=data,
               kind='bar')
#kind : bar, box, count, point,violin,strip
```

```python
sns.factorplot(x='drive-wheels',
               hue='num-of-doors',
               col='engine-location',
               data=data,
               kind='count')
#for kind=count, either x or y should be none
```

```python
#to draw linear regression plot
sns.lmplot(x='price', y='peak-rpm', data= data)
```
